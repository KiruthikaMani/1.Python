```python
 #3.Find Even and Odd Numbers in a List
Question: Write a Python program to separate even and odd numbers from the list [12, 17, 19, 24, 35, 40, 56].

```

```python
#3.ANS:
list =[12, 17, 19, 24, 35, 40, 56]
odd_list=[]
even_list=[]
for temp in list:
    if temp%2==0:
        even_list.append(temp)
    else:
        odd_list.append(temp)

print("Please find the Even_list:",even_list)
print("Please find the odd_list:",odd_list)
```

```python
#4. Reverse a Number
Question: Write a Python program to reverse the number 12345.

```

```python
#4.ANS:
def reverse_order(num):
    new=str(num)
    return new[::-1]

reverse_order(12345)
    
```

```python
#5. Find the Second Largest Number in a List
Question: Write a Python program to find the second largest number in the list [12, 45, 98, 34, 23].
```

```python
#5.ANS
def second_largest_number(numbers):
    list=set(numbers)
    max_number=max(list)
    print("This is the first largest number:",max_number)
    new_list=[]
    for t in list:
        if max_number!=t:
            new_list.append(t)
            #print(new_list)
    print("This is the second largest number:",end=' ')
    return max(new_list)
```

```python
second_largest_number((12, 45, 98, 34, 23))
```
