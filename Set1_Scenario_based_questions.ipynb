5.	Scenario: A text-processing tool reverses a given word or sentence for formatting purposes.
 Write logic to take a word or sentence as input and produce its reversed version.


```python
input_sentence=input("Enter the word or sentence:")
reversed_word = ""
for change in input_sentence:
    reversed_word = change + reversed_word
print(reversed_word)
```

7.	Scenario: A retail store offers a 20% discount if a customer’s total order exceeds $100. Write logic to calculate the final amount to be paid after applying the discount.

```python
Total_amount=int(input("Enter the Total amount:"))
if(Total_amount>100):
    offer=Total_amount*0.20
    print("Offer discount:",offer)
    print("Final amount:",Total_amount-offer)
else:
    print("Sorry!!..You are not eligible for our offer discount")
```

8.	Scenario: A banking system processes withdrawal requests and ensures the user has enough balance.
    Write logic to check if a user has enough balance before allowing a withdrawal and update the remaining balance accordingly.


```python
#Example
Balance_amount=int(input("Enter the balance amount:"))
Withdrawal_amount=int(input("Enter the withdrawal amount:"))
if(Withdrawal_amount<Balance_amount):
    final_bal=Balance_amount-Withdrawal_amount
    print("Your withdrawal has been completed..!!")
    print("And your final balance amount is :",final_bal)
else:
    print("Really sorry!!.. your are not having enough balance to withdrawal, kindly enter less amount again!!")
```

9. A calendar system verifies whether a given year is a leap year based on standard leap year rules.
 Write logic to determine whether a given year is a leap year.


```python
year=int(input("Enter the year:"))
if((year % 400 == 0) or (year % 4 == 0 and year % 100 != 0)):
    print(year, "-This is leap year")
else:
    print(year, "-This is not a leap year")
```
