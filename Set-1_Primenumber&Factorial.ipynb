```python
#To find - Prime number
def p_number(num):
    if num<2:
        return False
    else:
        for i in range(2,int(num**0.5)+1):
            if num%i==0:
                return False
        return True
n=11        
if p_number(n):
    print(f"{n}- This is a Prime number")
else:
    print(f"{n}- This is not a Prime number")   
        
```

```python
#To find - Factorial 
def factorial(fn):
    result=1
    for i in range(1,fn+1):
        result *=i
    return result
print(factorial(4))
```
